```python
from studydata import NerrStudyData
import seaborn as sns
import folium
from folium.plugins import MarkerCluster
from folium import IFrame
import ipywidgets as widgets
from pandas_profiling import ProfileReport
import pandas as pd
import numpy as np
from jupyter_dashboard import StudyMap


sns.set()
```

```python
nerr = NerrStudyData()
```

```python
wq_df = nerr.get_wq_data()
nut_df = nerr.get_nut_data()
met_df = nerr.get_met_data()
stations_df = nerr.get_stations()
```

```python
# https://towardsdatascience.com/interactive-controls-for-jupyter-notebooks-f5c94829aee6
    
ALL = 'ALL'
# numpy array
def create_dropdown_options(array):
    unique = array.unique().tolist()
    unique.sort()
#     unique.insert(0, ALL)
    return unique


def display_dataframe(df, locations_df):
    options = create_dropdown_options(df.StationCode)
    station_options = widgets.Dropdown(
        options=options
    )
    
    options = ["All", "ProvisionalPlus and Historical", "Historical", "ProvisionalPlus"]
    quality_options = widgets.Dropdown(
        options=options
    )

    def filter_df_by_code(df, station_code, quality_code):        

        # Filter by quality_code first...
        def filter_by_qc(df, quality_code):
            if quality_code == "All":
                return df

            if quality_code == "ProvisionalPlus and Historical":
                mask = (df["Historical"] == "1") & (df["ProvisionalPlus"] == "1")                
            elif quality_code == "Historical":
                mask = (df["Historical"] == "1") & (df["ProvisionalPlus"] == "0")
            elif quality_code == "ProvisionalPlus":
                mask = (df["Historical"] == "0") & (df["ProvisionalPlus"] == "1")

            return df.loc[mask]

        
        # Filter by quality code
        df = filter_by_qc(df, quality_code) 

        # Filter by station, exclude columns of no interest...
        include_cols = df.select_dtypes(["datetime", "float"]).columns.tolist()
    
    
        mask = (df["StationCode"] == station_code)
        df = df.loc[mask, include_cols]
        
        # Filter locations by station code
        mask = (locations_df["StationCode"] == station_code)
        study_map = StudyMap.create_stations_map(locations_df.loc[mask])

        display(study_map)
        if not df.empty:            
            display("Generating Report.  Please Wait...")
            display(
                ProfileReport(
                    df,
                    title=f"Profile Report for {station_code}",
                    progress_bar=False
                ).to_widgets()
            )
        else:
            print("No data was found with the selected filters.")
            print("Unable to generate report.")

    widgets.interact(
        filter_df_by_code,
        df=widgets.fixed(df),
        station_code=station_options,
        quality_code=quality_options
    )
```
